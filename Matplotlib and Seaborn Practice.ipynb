```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([4, 3, 5, 6, 7])
plt.subplot(1, 2, 1)
plt.plot(x, y, marker = 'o', ms = 10, mec = 'yellow', mfc = 'red', color = 'pink', linewidth = 6, linestyle = 'dashed')
plt.grid()
plt.title('First')

x1 = np.array([ 2,  4, 6, 8])
y1 = np.array([100, 200, 300, 400])
plt.subplot(1, 2, 2)
plt.plot(x1, y1, marker = '*', ms = 10, mec = 'green', mfc = 'orange', color = 'black', linewidth = 6, linestyle = 'dotted')
plt.grid()
plt.title('second')
```

```python
# scatter plot
x = np.array([1, 2, 3, 4, 5])
y = np.array([4, 3, 5, 6, 7])
plt.subplot(1, 2, 1)
plt.scatter(x, y, marker = 'o', color = 'red')
plt.grid()
plt.title('First')

x1 = np.array([ 2,  4, 6, 8])
y1 = np.array([100, 200, 300, 400])
plt.subplot(1, 2, 2)
plt.scatter(x1, y1, marker = 'o', color = 'blue', cmap = 'viridis')
plt.grid()
plt.title('second')
plt.colorbar()
```

```python
# bar charts
x = np.array(['A', 'B', 'C', 'D'])
y = np.array([3, 8, 1, 10])
plt.bar(x, y, color = 'red')
plt.show()
```

```python
x = np.array(['A', 'B', 'C', 'D'])
y = np.array([3, 8, 1, 10])
plt.barh(x, y, color = 'green')
plt.show()
```

```python
x = np.array(['A', 'B', 'C', 'D'])
y = np.array([3, 8, 1, 10])
plt.bar(x, y, color = 'red', width = 0.5)
for i, v in enumerate(y):
    plt.text(i, v+0.2, str(v), color = 'black', ha = 'center')
plt.show()
```

```python
x = np.array(['A', 'B', 'C', 'D'])
y = np.array([3, 8, 1, 10])
plt.barh(x, y, color = 'green', height = 0.3)
for i, v in enumerate(y):
    plt.text(v + 0.2,  i, str(v), color = 'black', va = 'center')
plt.show()
```

```python
x = np.random.normal(170, 10, 250)
plt.hist(x)
plt.show()
```

```python
# pie chart
y = np.array([17, 25, 25, 15])
mylabels = ['Apples', 'Banana',  'Cherries', 'Orange']
myexplode = [0.2, 0, 0, 0]


plt.pie(y, labels = mylabels, startangle = 90, explode = myexplode, shadow = True)
plt.legend()
plt.show()
```
