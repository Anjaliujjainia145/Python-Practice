```python
fn = ['vivek', 'atul', 'deepak', 'tarun']
sn = ['pal', 'shukla', 'tiwari', 'yadav']
list(map(lambda x, y : x + " " + y, fn, sn))
```

```python
fn = ['vivek', 'atul', 'deepak', 'tarun']
sn = ['pal', 'shukla', 'tiwari', 'yadav']
result = []
for i , j in zip(fn , sn):
    result.append(i + " " + j)
result
```
