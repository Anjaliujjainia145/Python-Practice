```python
# create a function which separate the integer and non integer values from a list
def value(a):
    b = []
    c = []
    for i in a:
        if type(i) == int:
            b.append(i)
        else:
            c.append(i)
    return b , c
y = [1, 2, 3, 4, 5, 'a', 'b', 'c']
value(y)
```

```python
# create a user define function which will ask for key and value

def key_value(a):
    a = int(input('Enter the length of the dictionary: '))
    d = {}
    for i in range(0, a):
        k = input('Enter the key: ')
        v = int(input('Enter the value: '))
        d[k] = v
        print(d)
        for i, j in d.items():
            if j == max(d.values()):
                print(f'maximum key and value pair is {i , j}')
            if j == min(d.values()):
                print(f'minimum key and value pair is {i, j}')
key_value(a)
```

```python
class Grandfather:
    def a(self):
        print('this is the grandfather')
    def b(self):
        print('hello from grandfather')
class Father(Grandfather):
    def d(self):
        print('this is the father')
    def e(self):
        print('hello from father')
class Child(Father):
    def f(self):
        print('i am the child')
obj = Child()
obj.a()
obj.b()
obj.d()
obj.e()
obj.f()
```

```python
# hiarchical inheritance
class Animal:
    def sound(self):
        print('animal make sound')
class Dogsound(Animal):
    def bark(self):
        print('dog is barking')
class Catsound(Animal):
    def meow(self):
        print('cat is meowing')
a = Animal()
a.sound()
b = Dogsound()
b.bark()
c = Catsound()
c.meow()
```
