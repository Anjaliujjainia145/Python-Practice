```python
import pandas as pd
import numpy as np

# Creating a synthetic employee dataset
np.random.seed(42)
data = {
    'EmployeeID': range(101, 151),
    'Name': [f'Emp_{i}' for i in range(101, 151)],
    'Department': np.random.choice(['Sales', 'HR', 'IT', 'Finance', 'Marketing'], 50),
    'Age': np.random.randint(22, 60, size=50),
    'Salary': np.random.randint(30000, 120000, size=50),
    'JoiningDate': pd.date_range(start='2015-01-01', periods=50, freq='M')
}
df = pd.DataFrame(data)
df.head()
```

```python
# find the average salary of each department
df.groupby('Department')['Salary'].mean()
```

```python
# how many employees are older than 40 and earn more than 70000?
# df1[(df1['Salary'] > 25000) & (df1['Location' ] == 'Jaipur')]
df[(df["Age"] > 40) & (df['Salary'] > 70000)]
```

```python
# what is the maximum age per department?
# df.groupby('Department')['Salary'].mean()
df.groupby('Department')['Age'].max()

```

```python
# sort the dataframe by salary in descending order and return the top 5 earners?

# df1.groupby(['City'], as_index = False)['Sales'].sum().sort_values(by='Sales', ascending = False).head()
df.sort_values('Salary', ascending = False).head(5)
```

```python
# add a new column experience that stores number of years since joining (assume current year is 2025)

# df1['new_col'] = df1['Salary']+500
# df['Experience'] = df['JoiningDate'] - 2025
df['Experience'] = 2025 - df['JoiningDate'].dt.year
df[['Name', 'Experience']].head()
```

```python
# filter the data for employees who joined between 2016 and 2018
import numpy as np
# df[(df["Age"] > 40) & (df['Salary'] > 70000)]
df[(df['JoiningDate'] >= '2016-01-01') & (df['JoiningDate'] <= '2018-12-31' )]
```

```python
# find the department with the highest average salary

# df.groupby('Department')['Age'].max()
df.groupby('Department')['Salary'].max()

```

```python
# count of employees per department who have more than 5 years of experience?
# df1.groupby(['Product Category'], as_index = False).agg({'Sales': 'sum', 'Profit': 'sum'})

df.groupby(['Department'], as_index = False).agg({'Name': 'count', 'EmployeeID': 'count'})
```

```python
# replace all department to uppercase
df['Department'] = df['Department'].str.upper()
df
```

```python
# create a new column calles 'seniority' whose experience is greater than 8 senior else junior
df['Seniority'] = df['Experience'].apply(lambda a : 'Senior' if a > 8 else 'Junior')
df[['Seniority', 'Name']].head()
```
